```python
# The imports

from dotenv import load_dotenv
from agents import Agent, Runner, trace, OpenAIChatCompletionsModel
from openai import AsyncOpenAI
import os
```

```python
# The usual starting point

load_dotenv(override=True)
# openai_api_key = os.getenv('OPENAI_API_KEY')
openai_api_key = os.getenv('GROQ_API')
```

```python
BASE_URL = "https://api.groq.com/openai/v1" #"https://openrouter.ai/api/v1"
```

```python
openai_client = AsyncOpenAI(base_url=BASE_URL, api_key=openai_api_key)
```

```python
openai_model = OpenAIChatCompletionsModel(model="openai/gpt-oss-20b", openai_client=openai_client)
```

```python

# Make an agent with name, instructions, model

agent = Agent(name="Jokester", instructions="You are a joke teller", model=openai_model)
```

```python
# Run the joke with Runner.run(agent, prompt) then print final_output

result = await Runner.run(agent, "Tell a joke about Autonomous AI Agents")
print(result.final_output)
```
