# Task: Response Text, Content, and Raw Socket

## Problem Statement:
Write a Python program that sends a request to a web page and:
1. Prints the response text (HTML content).
2. Prints the response content (in bytes).
3. Gets and prints the raw socket response from the server.

## Steps:
1. Import the `requests` library to send an HTTP request to the specified URL.
2. Send a `GET` request to the URL.
3. Extract the response text using `response.text` and print it.
4. Extract the response content in bytes using `response.content` and print it.
5. Use the `requests` library's `Raw` response interface to access the raw socket response, and print it.


```python
import requests
```

```python
res = requests.get('https://www.google.com/')
```

```python
print("Response text of https://google.com/")
print(res.text)
print("\n==============================================================================")
print("\nContent of the said url:")
print(res.content)
print("\n==============================================================================")
print("\nRaw data of the said url:")
r = requests.get('https://api.github.com/events', stream = True)
print(r.raw)
print(r.raw.read(15))
```
